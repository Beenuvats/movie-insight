```python
# Select relevant columns
movies = movies[['movie_id','title',"genres",'overview','keywords','cast','crew']]
movies.head(1)
```

```python
# Fill missing overview with empty strings instead of dropping
movies['overview'] = movies['overview'].fillna('')
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]['genres']
```

```python
import ast
# Function to convert stringified lists of dictionaries to lists of names
def convert(obj):
    return [i["name"] for i in ast.literal_eval(obj)]
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["genres"]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Function to get top 3 cast members
def convert3(obj):
    return [i["name"] for i in ast.literal_eval(obj)[:3]]
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
# Function to fetch the director
def fetch_director(obj):
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            return [i["name"]]
    return []
```

```python
movies["crew"] = movies['crew'].apply(fetch_director)
```

```python
# Text processing for tags
def clean_text(text):
    return text.lower().replace(" ", "")
```

```python
# transform each movie's overview (which is a string) into a list of words.
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['genres'] = movies['genres'].apply(lambda x: [clean_text(i) for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [clean_text(i) for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [clean_text(i) for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [clean_text(i) for i in x])
```

```python
movies.head(3)
```

```python
from wordcloud import WordCloud
keywords = movies['keywords'].explode()  
keywords_text = ' '.join(keywords.dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(keywords_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axis
plt.show()
```

```python
# Combine text data into a single 'tags' column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
# Function for stemming
ps = PorterStemmer()
def stem(text):
    return " ".join([ps.stem(i) for i in text.split()])
```

```python
# Apply stemming
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Create the tf-idf matrix
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Calculate cosine similarity matrix
similarity = cosine_similarity(vectors)
```

```python
similarity[0].shape
```

```python
import pickle
# Function to recommend movies
def recommend(movie):
    if movie in movies['title'].values:
        movie_index = movies[movies['title'] == movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        for i in movie_list:
            print(movies.iloc[i[0]].title)
    else:
        print("No movie found")
```

```python
# Example recommendation
recommend("Avatar")
```

```python
# Example recommendation
recommend("Spectre")
```

```python
import random

def get_recommended_items(user_id, k=5):
    
    return [f"movie{random.randint(1, 10)}" for _ in range(k)]

def get_relevant_items(user_id):

    return [f"movie{random.randint(1, 10)}" for _ in range(5)]

# Precision and recall functions
def precision_at_k(recommended_items, relevant_items, k):
    recommended_items_at_k = recommended_items[:k]
    relevant_recommendations = set(recommended_items_at_k) & set(relevant_items)
    return len(relevant_recommendations) / k

def recall_at_k(recommended_items, relevant_items, k):
    recommended_items_at_k = recommended_items[:k]
    relevant_recommendations = set(recommended_items_at_k) & set(relevant_items)
    return len(relevant_recommendations) / len(relevant_items)

user_id = 1

# Fetch data
recommended_items = get_recommended_items(user_id, k=5)
relevant_items = get_relevant_items(user_id)

# Calculate precision and recall
p_at_k = precision_at_k(recommended_items, relevant_items, k=5)
r_at_k = recall_at_k(recommended_items, relevant_items, k=5)

print(f"Recommended Items: {recommended_items}")
print(f"Relevant Items: {relevant_items}")
print(f"Precision at 5: {p_at_k}")
print(f"Recall at 5: {r_at_k}")
```

```python
def f1_score(precision, recall):
    if (precision + recall) == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)
```

```python
ks = [3, 5, 7 ]
precisions = [precision_at_k(recommended_items, relevant_items, k) for k in ks]
recalls = [recall_at_k(recommended_items, relevant_items, k) for k in ks]

# Plot Precision-Recall Curve
plt.figure()
plt.plot(recalls, precisions, marker='o')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```
